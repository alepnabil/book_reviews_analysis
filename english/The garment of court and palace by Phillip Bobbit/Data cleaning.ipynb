# Data cleaning

Steps:

    
1. Remove duplicate reviews if there are duplicates



2. Convert text data to lowercase

4. Remove stop words




5. Standardizing text


6. Correcting spelling

Reading our data which we gathered.

```python
df_final=pd.read_csv('final dataframe.csv')
df_final
```

```python
df_final=df_final.drop(df_final.columns[0], axis = 1) 
```

```python
df_final
```

#### Remove '\n\n' from our reviews

```python
df_final['reviews']=df_final['reviews'].apply(lambda x: x.strip('\n'))
```

#### Removing duplicate reviews.

There are some reviews which was scraped twice and have duplicates

```python
df_final['reviews'][24]
```

Function to remove duplicates

```python
def remove_duplicate(column):
    new=column.split('\n')
    if len(new)>1:
        new1=new[1]
    else:
        new1=new[0]
    return new1
```

```python
df_final['reviews']=df_final['reviews'].apply(remove_duplicate)
```

```python
df_final['reviews'][24]
```

#### Converting reviews text to lower case

```python
df_final['reviews']=df_final['reviews'].apply(lambda x:x.lower())
```

#### Removing punctuation

```python
df_final['reviews']=df_final['reviews'].apply(lambda x:x.lower())
```

#### Remove stop words

```python
stop= stopwords.words('english')
```

```python
df_final['reviews']=df_final['reviews'].apply(lambda x : ' '.join(review for review in x.split() if review not in stop))
```

```python
df_final['reviews'][24]
```

#### Correcting spelling

```python
df_final['reviews']=df_final['reviews'].apply(lambda x:str(TextBlob(x).correct()))
```
