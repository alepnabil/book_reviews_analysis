### Getting reviews

Go to the first page----->Scrape reviews--->click on next page----->scrape reviews on second page

```python
path='C:\\Program Files (x86)\\chromedriver.exe'
driver=webdriver.Chrome(path)
```

```python
driver=webdriver.Chrome(path)

#redirect to the page
driver.get('https://www.amazon.com/Garments-Court-Palace-Machiavelli-World/product-reviews/0802120741/ref=cm_cr_getr_d_paging_btm_prev_1?ie=UTF8&reviewerType=all_reviews&pageNumber=1')
soup=BeautifulSoup(driver.page_source,'html.parser')

#scrape all the reviews
first_page_review=soup.select('div.a-row.a-spacing-small.review-data')

#print the reviews
for x in first_page_review:
    print(x.text)
    print('----')

time.sleep(10)

#find the next page button
next_page=driver.find_element_by_xpath('//*[@id="cm_cr-pagination_bar"]/ul/li[2]/a')
#click on that button
next_page.click()

driver.refresh()


#find the reviews
second_page_review=soup.select('span.a-size-base.review-text.review-text-content')


print('-----------SECOND PAGE REVIEWS-------')
for y in second_page_review:
    print(y.text)
    print('----')

    
time.sleep(7)

driver.quit()
```

This solution goes to the second page but it scrapes the review from the first page. This is maybe due to the reviews on the second and first having the same elements?

**solution**

So, compared to the first script, this script opens up 2 browsers (page 1 and page 2) then scrape each time they go the website.

```python
driver=webdriver.Chrome(path)
driver2=webdriver.Chrome(path)


#redirect to the page
driver.get('https://www.amazon.com/Garments-Court-Palace-Machiavelli-World/product-reviews/0802120741/ref=cm_cr_getr_d_paging_btm_prev_1?ie=UTF8&reviewerType=all_reviews&pageNumber=1')
soup=BeautifulSoup(driver.page_source,'html.parser')

#scrape all the reviews
first_page_review=soup.select('div.a-row.a-spacing-small.review-data')

#print the reviews
for x in first_page_review:
    print(x.text)
    print('----')
    reviews.append(x.text)
      
#delay
time.sleep(7)

#quit the first chrome browser
driver.quit()

#delay
time.sleep(7)


#open up second chrome browser
driver2.get('https://www.amazon.com/Garments-Court-Palace-Machiavelli-World/product-reviews/0802120741/ref=cm_cr_getr_d_paging_btm_prev_1?ie=UTF8&reviewerType=all_reviews&pageNumber=2')
soup2=BeautifulSoup(driver2.page_source,'html.parser')

#get the reviews
second_page_review=soup2.select('span.a-size-base.review-text.review-text-content')


print('-----------SECOND PAGE REVIEWS-------')
for y in second_page_review:
    print(y.text)
    print('----')
    reviews.append(y.text)

    
time.sleep(7)

driver2.quit()
```

```python
reviews=pd.Series(reviews)
```

### Getting user's username

```python
driver=webdriver.Chrome(path)
driver2=webdriver.Chrome(path)


#redirect to the page
driver.get('https://www.amazon.com/Garments-Court-Palace-Machiavelli-World/product-reviews/0802120741/ref=cm_cr_getr_d_paging_btm_prev_1?ie=UTF8&reviewerType=all_reviews&pageNumber=1')
soup=BeautifulSoup(driver.page_source,'html.parser')

#scrape all the reviews
first_page_username=soup.select('span.a-profile-name')

#print the reviews
for x in first_page_username:
    print(x.text)
    print('----')
    username.append(x.text)
      
#delay
time.sleep(7)

#quit the first chrome browser
driver.quit()

#delay
time.sleep(7)


#open up second chrome browser
driver2.get('https://www.amazon.com/Garments-Court-Palace-Machiavelli-World/product-reviews/0802120741/ref=cm_cr_getr_d_paging_btm_prev_1?ie=UTF8&reviewerType=all_reviews&pageNumber=2')
soup2=BeautifulSoup(driver2.page_source,'html.parser')

#get the reviews
second_page_username=soup2.select('span.a-profile-name')


print('-----------SECOND PAGE REVIEWS-------')
for y in second_page_username:
    print(y.text)
    print('----')
    username.append(y.text)

    
time.sleep(7)

driver2.quit()
```

```python
username=pd.Series(username)
```

Seems here there are 2 duplicates from second page reviewers username. Need to drop the duplicates. Then rearrange the index

```python
username=username.drop_duplicates(keep='first')
```

```python
username=username.reset_index(drop=True)
```

# Scraping from goodreads

```python
goodread_link='https://www.goodreads.com/en/book/show/15815812-the-garments-of-court-and-palace'

goodread_request=requests.get(goodread_link)


Soup=BeautifulSoup(goodread_request.content,'html.parser')
```

```python
goodread_user=Soup.select('a.user')
goodread_user
```

```python
for x in goodread_user:
    print(x.text)
```

```python
goodread_user=[x.text for x in goodread_user]
goodread_user=pd.Series(goodread_user)
goodread_user
```

scrape user's reviews

```python
goodread_reviews=Soup.select('div.reviewText.stacked')

goodread_reviews=[x.text for x in goodread_reviews]

goodread_reviews=pd.Series(goodread_reviews)


goodread_reviews
```

```python
df_goodread=pd.concat([goodread_user,goodread_reviews],axis=1)
df_goodread.columns=['username','reviews']
df_goodread
```

# Final dataframe

```python
df_final=pd.concat([df,df_goodread],axis=0)
df_final
```
