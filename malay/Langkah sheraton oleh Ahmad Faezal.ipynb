# Preprocessing text

1.convert words to lower case

2.remove punctuations

3.spellling correction


The reviews are somehow repeated. Therefore we need to clean the data and get the correct reviews we want.

```python
df['reviews'][0]
```

```python
df['reviews'][0].split('\n')[1]
```

Making a function that removes reviews with duplicates.

```python
def remove_duplicate(column):
    new=column.split('\n')
    if len(new)>1:
        new1=new[1]
    else:
        new1=new[0]
    return new1
```

```python
df['reviews']=df['reviews'].apply(remove_duplicate)
```

```python
df['reviews'][0]
```

Now, our reviews does not have duplicates.

### Converting words to lower case

```python
df['reviews']=df['reviews'].apply(lambda x:x.lower())
```

### Removing punctuation

```python
df['reviews']=df['reviews'].str.replace('[^\w\s]','')
```

### Spelling correction

Credit to the creator and contributers of Malaya:https://malaya.readthedocs.io/en/latest/load-spell-correction.html

```python
word_corrector=malaya.spell.probability()
```

```python
df['reviews']=df['reviews'].apply(lambda x:word_corrector.correct_text(x))
```

```python
df['reviews'].str.split(expand=True).stack()
```

```python
word_frequencies=pd.DataFrame(df['reviews'].str.split(expand=True).stack().value_counts())
word_frequencies[:30].style.background_gradient(cmap='Blues')
```

```python
plt.figure(figsize=(12,8))
word_frequency[:30].plot(kind='barh',color='violet')
plt.xlabel('Word frequency')
plt.ylabel('Words')
```

# Segment reviews based on their sentiment

```python
import random
```

```python
positive_reviews=df[df['sentiment']=='positive']
negative_reviews=df[df['sentiment']=='negative']
neutral_reviews=df[df['sentiment']=='neutral']
```

```python
positive_reviews
```

```python
colour=['green','yellow','red']
labels=['Positive reviews','Negative reviews','Neutral reviews']


def words_frequency(dataframe):
    word_frequency=pd.Series(dataframe['reviews']).str.split(expand=True).stack().value_counts()
    word_frequency
   
    
    plt.figure(figsize=(12,8))
    word_frequency[:30].plot(kind='barh',color=random.choice(colour))
    plt.xlabel('Word frequency')
    plt.ylabel('Words')
    
    temp_review=''.join(dataframe['reviews'])
    
    plt.figure(figsize=(12,6))
    wordcloud=WordCloud(background_color='lightblue').generate(temp_review)
    plt.imshow(wordcloud,interpolation='bilinear')
```

# Positive reviews

```python
words_frequency(positive_reviews)
```

Here we can see words like 'bruh' got corrected into 'buruh'. Canot say much about the positive reviews since the reviews aren't that much. Maybe the solution to this problem maybe solved by collecting more data(reviews) from other platforms such as social medias.
